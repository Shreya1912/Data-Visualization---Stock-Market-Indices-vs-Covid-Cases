```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import cufflinks as cf
init_notebook_mode(connected=True)
cf.go_offline()
import plotly.express as px
from pandas_datareader import data, wb
from plotly.offline import download_plotlyjs,init_notebook_mode,plot
import plotly.graph_objs as go
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from functools import reduce

#Setting display rows to Max
pd.set_option('display.max_rows', None)
```

```python
#Reading covid_data from 31 Dec 2019 to 10 July 2020
covid_data = pd.read_csv("owid-covid-data-10July.csv")
```

```python
#Extraxting covid data specific to US
US_data=covid_data[covid_data['iso_code']=='USA']
```

```python
#Reading the US dataframe
US_data.head()
```

```python
#Plotting total cases vs new cases

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(x=US_data['date'], y=US_data['new_cases'], name="New Cases"),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=US_data['date'], y=US_data['total_cases'], name="Total Cases"),
    secondary_y=True,
)

# Add figure title
fig.update_layout(
    title={
        'text': "Analysing Covid Cases in the US",
        'y':0.98,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})

# Set x-axis title
fig.update_xaxes(title_text="Date")

# Set y-axes titles
fig.update_yaxes(title_text="New Csses", secondary_y=False)
fig.update_yaxes(title_text="Total Cases", secondary_y=True)


fig.update_xaxes(showgrid=False, zeroline=True, rangeslider_visible=True, showticklabels=True,
                 showspikes=True, spikemode='across', spikesnap='cursor', showline=False,
                 spikecolor="grey",spikethickness=0.5, spikedash='solid')

fig.update_yaxes(showgrid=False,showspikes=True, spikedash='solid',spikemode='across', 
                spikecolor="grey",spikesnap="cursor",spikethickness=0.5)
 
fig.update_layout({'plot_bgcolor': "#21201f", 'paper_bgcolor': "#21201f", 'legend_orientation': "h"},
                  legend=dict(y=1, x=0),
                  font=dict(color='#dedddc'), dragmode='pan', hovermode='x unified',
                  margin=dict(b=20, t=0, l=0, r=40))

fig.show() 
```

```python
#Making main data frame with only required columns

main_data = US_data[['date','total_cases','new_cases','total_deaths']]
main_data['date'] = pd.to_datetime(main_data['date'])
main_data
```

```python
#Reading Dow Jones and S&P 500 Index Using Yahoo Finance
start = '2019-12-31'
end = '2020-07-10'

# S&P 500 9Ticker: INX
SNP = data.DataReader('^GSPC', 'yahoo', start, end)

# Dow Jones Industrial Average: DJI
DJI = data.DataReader('^DJI', 'yahoo', start, end)

# Volatility Index: VIX
VIX = data.DataReader('^VIX', 'yahoo', start, end)

# Tech Sector ETF: IGM
IGM = data.DataReader('IGM', 'yahoo', start, end)

# Oil Sector ETF: IGM
XLE = data.DataReader('XLE', 'yahoo', start, end)

#United States Oil Fund: USO
USO = data.DataReader('USO', 'yahoo', start, end)

#Gold Shares: GLD
GLD = data.DataReader('GLD', 'yahoo', start, end)
```

```python
#Renaming the column 'Close' so it becomes identifiable, dropping columns that are not required

SNP.rename(columns={'Close':'SNP Close'}, inplace=True)
SNP['date'] = SNP.index
SNP.drop(['High','Low','Open','Volume','Adj Close'], axis=1, inplace=True)

DJI.rename(columns={'Close':'DJI Close'}, inplace=True)
DJI['date'] = DJI.index
DJI.drop(['High','Low','Open','Volume','Adj Close'], axis=1, inplace=True)

VIX.rename(columns={'Close':'VIX Close'}, inplace=True)
VIX['date'] = VIX.index
VIX.drop(['High','Low','Open','Volume','Adj Close'], axis=1, inplace=True)

IGM.rename(columns={'Close':'IGM Close'}, inplace=True)
IGM['date'] = IGM.index
IGM.drop(['High','Low','Open','Volume','Adj Close'], axis=1, inplace=True)

XLE.rename(columns={'Close':'XLE Close'}, inplace=True)
XLE['date'] = XLE.index
XLE.drop(['High','Low','Open','Volume','Adj Close'], axis=1, inplace=True)

USO.rename(columns={'Close':'USO Close'}, inplace=True)
USO['date'] = USO.index
USO.drop(['High','Low','Open','Volume','Adj Close'], axis=1, inplace=True)

GLD.rename(columns={'Close':'GLD Close'}, inplace=True)
GLD['date'] = GLD.index
GLD.drop(['High','Low','Open','Volume','Adj Close'], axis=1, inplace=True)
```

```python
GLD
```

```python
#Merging all the frames togather
data_frames = [SNP, DJI, VIX, IGM, XLE, USO, GLD]
main_data3 = pd.concat(data_frames, join='outer', axis=1)
main_data4 = main_data3.loc[:,~main_data3.columns.duplicated()]
main_data4.info()
```

```python
#Merging Covid data and stock indexes
main_data5=pd.merge(main_data, main_data4, how='outer', on='date')

#Identifying null values
main_data5.info()

#Null values in stock indexes are due to holiday
```

```python
#Filling null values with one day previous value
main_data6=main_data5.fillna(method='ffill')
main_data6.info()
```
